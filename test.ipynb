```python
import cv2
from matplotlib import pyplot as plt


# load an image in grayscale mode
img = cv2.imread('testimag1.jpg',0)
  
# calculate frequency of pixels in range 0-255
histg = cv2.calcHist([img],[0],None,[256],[0,256]) 
plt.plot(histg)
plt.show()
```
